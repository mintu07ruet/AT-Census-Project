# combine all year Strava volume and calculate the market penetration

```python
mp19=d19[['id','lat','long','osmId','edge_uid','stv_total','stv_rec_to','stv_com_to']]
mp19=mp19.rename(columns={'stv_total':'stv19'})
mp19=mp19.rename(columns={'stv_rec_to':'stv_rec19'})
mp19=mp19.rename(columns={'stv_com_to':'stv_com19'})
#mp19['average19stv']=mp19['stv_total'].mean()
#mp19['average19stv_rec']=mp19['stv_rec_to'].mean()
#mp19['average19stv_com']=mp19['stv_com_to'].mean()
mp19=mp19.drop_duplicates(subset=['id'],keep='first')
#mp19=mp19[['id','lat','long','osmId','edge_uid','average19stv','average19stv_rec','average19stv_com']]
mp19.head()
```

```python
mp18=eco_stv18[['id','lat','long','osmId','edge_uid','stv_total','stv_rec_total','stv_com_total']]
mp18=mp18.rename(columns={'stv_total':'stv18'})
mp18=mp18.rename(columns={'stv_rec_total':'stv_rec18'})
mp18=mp18.rename(columns={'stv_com_total':'stv_com18'})
mp18=mp18.drop_duplicates(subset=['id'],keep='first')
#mp18=mp18[['id','lat','long','osmId','edge_uid','average18stv','average18stv_rec','average18stv_com']]
mp18.head()
```

```python
mp20=eco_stv20[['id','lat','long','osmId','edge_uid','stv_total','stv_rec_total','stv_com_total']]
mp20=mp20.rename(columns={'stv_total':'stv20'})
mp20=mp20.rename(columns={'stv_rec_total':'stv_rec20'})
mp20=mp20.rename(columns={'stv_com_total':'stv_com20'})
mp20=mp20.drop_duplicates(subset=['id'],keep='first')
mp20.head()
```

```python
mp21=eco_stv21[['id','lat','long','osmId','edge_uid','stv_total','stv_rec_total','stv_com_total']]
mp21=mp21.rename(columns={'stv_total':'stv21'})
mp21=mp21.rename(columns={'stv_rec_total':'stv_rec21'})
mp21=mp21.rename(columns={'stv_com_total':'stv_com21'})
mp21=mp21.drop_duplicates(subset=['id'],keep='first')
mp21.head()
```

```python
mp1819=pd.merge(mp18,mp19, left_on=['id','lat','long','osmId','edge_uid'],right_on=['id','lat','long','osmId','edge_uid'],how='inner')
mp181920=pd.merge(mp1819,mp20, left_on=['id','lat','long','osmId','edge_uid'],right_on=['id','lat','long','osmId','edge_uid'],how='inner')
mp18192021=pd.merge(mp181920,mp21, left_on=['id','lat','long','osmId','edge_uid'],right_on=['id','lat','long','osmId','edge_uid'],how='inner')
mp18192021.head()
```

```python
mp18192021['stv18_ave']=mp18192021['stv18'].mean()
mp18192021['stv19_ave']=mp18192021['stv19'].mean()
mp18192021['stv20_ave']=mp18192021['stv20'].mean()
mp18192021['stv21_ave']=mp18192021['stv21'].mean()
mp18192021.head()
```

```python
mp18192021['stv_com18_ave']=mp18192021['stv_com18'].mean()
mp18192021['stv_com19_ave']=mp18192021['stv_com19'].mean()
mp18192021['stv_com20_ave']=mp18192021['stv_com20'].mean()
mp18192021['stv_com21_ave']=mp18192021['stv_com21'].mean()
```

```python
mp18192021['stv_rec18_ave']=mp18192021['stv_rec18'].mean()
mp18192021['stv_rec19_ave']=mp18192021['stv_rec19'].mean()
mp18192021['stv_rec20_ave']=mp18192021['stv_rec20'].mean()
mp18192021['stv_rec21_ave']=mp18192021['stv_rec21'].mean()
mp18192021.head()
```

```python
mp18192021['mp19stv']=((mp18192021['stv19']-mp18192021['stv18'])/mp18192021['stv18'])*100
mp18192021['mp19stv_com']=((mp18192021['stv_com19']-mp18192021['stv_com18'])/mp18192021['stv_com18'])*100
mp18192021['mp19stv_rec']=((mp18192021['stv_rec19']-mp18192021['stv_rec18'])/mp18192021['stv_rec18'])*100
```

```python
mp18192021['mp20stv']=((mp18192021['stv20']-mp18192021['stv19'])/mp18192021['stv19'])*100
mp18192021['mp20stv_com']=((mp18192021['stv_com20']-mp18192021['stv_com19'])/mp18192021['stv_com19'])*100
mp18192021['mp20stv_rec']=((mp18192021['stv_rec20']-mp18192021['stv_rec19'])/mp18192021['stv_rec19'])*100
```

```python
mp18192021['mp21stv']=((mp18192021['stv21']-mp18192021['stv20'])/mp18192021['stv20'])*100
mp18192021['mp21stv_com']=((mp18192021['stv_com21']-mp18192021['stv_com20'])/mp18192021['stv_com20'])*100
mp18192021['mp21stv_rec']=((mp18192021['stv_rec21']-mp18192021['stv_rec20'])/mp18192021['stv_rec20'])*100
```
