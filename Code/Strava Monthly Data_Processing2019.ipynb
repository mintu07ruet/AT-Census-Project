```python
# Join shape file with data based edgeUID and OSM id
strava_data=pd.merge(CA_Strava_data,CA_Strava,left_on=['edge_uid','osm_reference_id'],right_on=['edgeUID','osmId'], how='inner')
strava_data.shape
```

```python
strava_data['stv_total']=strava_data['forward_trip_count']+strava_data['reverse_trip_count']
strava_data['stv_com_total']=strava_data['forward_commute_trip_count']+strava_data['reverse_commute_trip_count']
strava_data['stv_rec_total']=strava_data['forward_leisure_trip_count']+strava_data['reverse_leisure_trip_count']
strava_data['stv_com_ratio']=strava_data['stv_com_total']/strava_data['stv_total']
strava_data['stv_rec_ratio']=strava_data['stv_rec_total']/strava_data['stv_total']
strava_data.head()
```
